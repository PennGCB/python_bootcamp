```python
column_names = [
    "peak_chr",   # chromosome where peak is at
    "peak_start", # peak start site
    "peak_stop",  # peak stop site
    "peak_name", # name of peak
    "peak_signal", # signal score of peak
    "peak_strand", # strand where peak is found 
                   #(its always strandless)
    "gene_chr", # chromosome where gene is at
    "gene_start", # gene start site
    "gene_stop", # gene stop site
    "gene_name", # name of gene
    "gene_signal", # gene signal score
    "gene_strand", # strand where gene is found
    "distance" # distance from peak to gene
]
```

```python
column_types = {"peak_chr" : str,
                "peak_start" : np.int64,
                "peak_stop" : np.int64,
                "peak_name" : str,
                "peak_signal" : np.float64,
                "peak_strand":str,
                "gene_chr" : str,
                "gene_start" : np.int64,
                "gene_stop" : np.int64,
                "gene_name" : str, 
                "gene_signal" : np.float64,
                "gene_strand" : str,
                "distance" : np.float64}
```
