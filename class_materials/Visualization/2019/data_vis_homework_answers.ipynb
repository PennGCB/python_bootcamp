# 1.  Importance of Data Visualization

**(A)** Calculate the mean of x, the mean of y, and the regression coefficients of each (x,y) pair.

**Hint:** The `stats.linregress` function from scipy may be useful here

```python
import random
import string

import pandas
from plotnine import *
import numpy as np
import scipy.stats

# RUN ME FIRST
x1 = [10, 8, 13, 9, 11, 14, 6, 4, 12, 7, 5]
y1 = [8.04, 6.95, 7.58, 8.81, 8.33, 9.96, 7.24, 4.26, 10.84, 4.82, 5.68]

x2 = [10, 8, 13, 9, 11, 14, 6, 4, 12, 7, 5]
y2 = [9.14, 8.14, 8.74, 8.77, 9.26, 8.10, 6.13, 3.10, 9.13, 7.26, 4.74]

x3 = [10, 8, 13, 9, 11, 14, 6, 4, 12, 7, 5]
y3 = [7.46, 6.77, 12.74, 7.11, 7.81, 8.84, 6.08, 5.39, 8.15, 6.42, 5.73]
x4 = [8, 8, 8, 8, 8, 8, 8, 19, 8, 8, 8]
y4 = [6.58, 5.76, 7.71, 8.84, 8.47, 7.04, 5.25, 12.50, 5.56, 7.91, 6.89]
```

```python
def mean(x):
    return sum(x) / len(x)

print(mean(x1))
print(mean(x2))
print(mean(x3))
print(mean(x4))

print(mean(y1))
print(mean(y2))
print(mean(y3))
print(mean(y4))

print(scipy.stats.linregress(x1, y1).slope, scipy.stats.linregress(x1, y1).intercept)
print(scipy.stats.linregress(x2, y2).slope, scipy.stats.linregress(x2, y2).intercept)
print(scipy.stats.linregress(x3, y3).slope, scipy.stats.linregress(x3, y3).intercept)
print(scipy.stats.linregress(x4, y4).slope, scipy.stats.linregress(x4, y4).intercept)
```

# 2. Heatmap

**(A)** Create a pandas dataframe from the data stored in heatmap.csv

```python
# create heatmap.csv
num_genes = 30
genes = []
for i in range(num_genes):
    gene_base = ''.join([random.choice(string.ascii_uppercase) for _ in range(3)])
    genes.append(gene_base + str(random.choice([0,1,2,3,4])))

case_1 = np.random.normal(size=num_genes) + 1.5
case_2 = np.random.normal(size=num_genes) + 1.7
case_3 = np.random.normal(size=num_genes) + 1.6
control_1 = np.random.normal(size=num_genes)
control_2 = np.random.normal(size=num_genes)

    
heatmap_df = pandas.DataFrame({'genes': genes, 'case_1': case_1, 'case_2':case_2,
                               'case_3': case_3, 'control_1': control_1, 'control_2': control_2})

heatmap_df.to_csv('heatmap.csv', index=False)
```

```python
heatmap_df = pandas.read_csv('heatmap.csv')

heatmap_df = heatmap_df.melt(id_vars=['genes'])
heatmap_df = heatmap_df.rename(columns={'variable': 'case/control', 'value': 'expression_value'})

heatmap_df.head()
```

**(B)** Create a heatmap from the dataframe

```python
ggplot(heatmap_df, aes(x='case/control', y='genes'))\
+ geom_tile(aes(fill='expression_value'))\
```

**(C)** Change the heatmap to a gradient of your favorite color

```python
ggplot(heatmap_df, aes(x='case/control', y='genes'))\
+ geom_tile(aes(fill='expression_value'))\
+ scale_fill_gradient('blue')
```

**(D)** Add the values of each element in the heatmap as text

```python
ggplot(heatmap_df, aes(x='case/control', y='genes'))\
+ geom_tile(aes(fill='expression_value'))\
+ geom_label(aes(label='expression_value'), size=5)
```

# 3. Box Plot

**(A)** Create a dataframe from the data in box_plot.tsv

```python
# Generate Box Plot Data

control = np.random.normal(size=50) * 5 + 130
case1 = np.random.normal(size=20) * 5 + 150
case2 = np.random.normal(size=30) * 5 + 90

case = np.append(case1, case2)

print(control.shape)
print(case.shape)

df = pandas.DataFrame({'treated': case, 'control': control})
df.to_csv('box_plot.tsv', sep='\t', index=None)
```

```python
box_df = pandas.read_csv('box_plot.tsv', sep='\t')

box_df = box_df.melt()

box_df = box_df.rename(columns={'variable': 'treated/control', 'value': 'blood_pressure'})

box_df.head()
```

**(B)** Create a box plot comparing the blood pressure of the case and control groups

```python
ggplot(box_df, aes(x='treated/control', y='blood_pressure', fill='treated/control')) + geom_boxplot()
```

**(C)** Create a violin plot showing the same data

```python
ggplot(box_df, aes(x='treated/control', y='blood_pressure', fill='treated/control')) + geom_violin()
```

**(D)** What key feature of the data can you see in the violin plot that is hidden in the box plot?

The distribution of treated individuals is bimodal. This means that while the treatment improved the average patient, 40% of the patients are actually worse off for having taken the drug.
